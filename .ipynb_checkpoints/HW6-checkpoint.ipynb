# (a)  Download the Breast Cancer Wisconsin (Diagnostic) Data Set from: https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+ %28Diagnostic%29. Download the data in https://archive.ics.uci.edu/ml/ machine-learning-databases/breast-cancer-wisconsin/wdbc.data, which has IDs, classes (Benign=B, Malignant=M), and 30 attributes. This data has two output classes. Use the first 20% of the positive and negative classes in the file as the test set and the rest as the training set

```python
import csv
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("./data/wdbc.data", header=None)
data = data.iloc[:, 1:]
benign_num = data.iloc[:, 0].value_counts()['B']*.2
malignant_num = data.iloc[:, 0].value_counts()['M']*.2

benign_count = 0
malignant_count = 0
train_data = []
test_data = []
for line in data.values:
    if line[0] == 'B':
        line[0] = 1
        if benign_count <= benign_num:
            test_data.append(line)
        else:
            train_data.append(line)
        benign_count += 1
    else:
        line[0] = 0
        if malignant_count <= malignant_num:
            test_data.append(line)
        else:
            train_data.append(line)
        malignant_count += 1

train_data = pd.DataFrame(train_data)
test_data = pd.DataFrame(test_data)
train_data.to_csv("./src/train_data.csv", index=False, header=False)
test_data.to_csv("./src/test_data.csv", index=False, header=False)
all_data = train_data.append(test_data)
all_data.to_csv("./src/all_data.csv", index=False, header=False)
```

## ii. Semi-Supervised Learning/ Self-training: select 50% of the positive class along with 50% of the negative class in the training set as labeled data and the rest as unlabelled data. You can select them randomly

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score, auc

warnings.filterwarnings("ignore")
data = pd.read_csv("./src/all_data.csv", header=None)
train_data_, test_data, train_target_, test_target = train_test_split(normalize(data.iloc[:, 1:]), data.iloc[:, 0],
                                                                      test_size=0.2, stratify=data.iloc[:, 0])
train_accuracy = 0
test_accuracy = 0
train_precision = 0
test_precision = 0
train_recall = 0
test_recall = 0
train_f1_score = 0
test_f1_score = 0
train_auc = 0
test_auc = 0
for i in range(30):
    # train SVM using labeled data
    label_data, unlabel_data, label_target, unlabel_target = train_test_split(train_data_, train_target_,
                                                                              test_size=0.5, stratify=train_target_)

    label_data = pd.DataFrame(label_data)
    unlabel_data = pd.DataFrame(unlabel_data)

    SVCParameters = {
        "tol": np.logspace(-5, 2, 7), "C": np.linspace(0.01, 5, 10)}
    clf = LinearSVC(penalty='l1', dual=False)
    CV = GridSearchCV(clf, SVCParameters, cv=5)
    CV.fit(label_data, label_target)
    best_clf = CV.best_estimator_

    # Use SVM to label unlabeled data
    while len(unlabel_data) > 0:

        dist = best_clf.decision_function(
            unlabel_data)
        index = np.abs(dist).argmax()

        label_data = label_data.append(
            unlabel_data.iloc[index, :], ignore_index=True)
        p = best_clf.predict(unlabel_data.iloc[index, :].values.reshape(1, -1))
        label_target = np.append(label_target, p)
        unlabel_data = unlabel_data.drop(index)
        unlabel_data = unlabel_data.reset_index(drop=True)
        best_clf.fit(label_data, label_target)

    train_data = label_data
    train_target = label_target
    train_pred = best_clf.predict(train_data)
    train_accuracy += accuracy_score(train_target, train_pred)
    train_precision += precision_score(
        train_target, train_pred)
    train_recall += recall_score(train_target, train_pred)
    train_f1_score += f1_score(train_target, train_pred)
    train_fpr, train_tpr, thresholds = roc_curve(train_target, train_pred)
    train_auc += auc(train_fpr, train_tpr)

    test_pred = best_clf.predict(test_data)

    test_accuracy += accuracy_score(test_target, test_pred)
    test_precision += precision_score(
        test_target, test_pred)
    test_recall += recall_score(test_target, test_pred)
    test_f1_score += f1_score(test_target, test_pred)
    test_fpr, test_tpr, thresholds = roc_curve(test_target, test_pred)
    test_auc += auc(test_fpr, test_tpr)

print('avg train accuracy =', train_accuracy/30)
print('avg train precision =', train_precision/30)
print('avg train recall =', train_recall/30)
print('avg train f1_score =', train_f1_score/30)
print('avg train auc =', train_auc/30)
print('Confusion Matrix of training in last run ')
print(confusion_matrix(train_target, train_pred))

print('avg test accuracy =', test_accuracy/30)
print('avg test precision =', test_precision/30)
print('avg test recall =', test_recall/30)
print('avg test f1_score =', test_f1_score/30)
print('avg test auc =', test_auc/30)
print('Confusion Matrix of test in last run ')
print(confusion_matrix(test_target, test_pred))

plt.plot(train_fpr, train_tpr, label='train ROC curve')
plt.plot(test_fpr, test_tpr, label='test ROC curve')
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

## iii. Unsupervised Learning: Run k-means algorithm on the whole training set. Ignore the labels of the data, and assume k = 2.



```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

data = pd.read_csv("./src/all_data.csv", header=None)

train_accuracy = 0
test_accuracy = 0
train_precision = 0
test_precision = 0
train_recall = 0
test_recall = 0
train_f1_score = 0
test_f1_score = 0
train_auc = 0
test_auc = 0
for i in range(30):
    train_data, test_data, train_target, test_target = train_test_split(normalize(data.iloc[:, 1:]), data.iloc[:, 0],
                                                                        test_size=0.2, stratify=data.iloc[:, 0])
    clf = KMeans(n_clusters=2, init='random', n_init=40)
    clf.fit(train_data)
    train_pred = clf.predict(train_data)
    dist_1 = cdist(train_data, clf.cluster_centers_)[:, 0]
    dist_2 = cdist(train_data, clf.cluster_centers_)[:, 1]
    closest_index1 = np.argsort(dist_1, axis=0)[:30]
    closest_index2 = np.argsort(dist_2, axis=0)[:30]

    cluster1_label = np.argmax(np.bincount(
        train_target.iloc[closest_index1]))
    cluster2_label = np.argmax(np.bincount(
        train_target.iloc[closest_index2]))

    for j in range(len(train_pred)):
        if train_pred[j] == 0:
            train_pred[j] = cluster1_label
        else:
            train_pred[j] = cluster2_label

    train_accuracy += accuracy_score(train_target, train_pred)
    train_precision += precision_score(
        train_target, train_pred)
    train_recall += recall_score(train_target, train_pred)
    train_f1_score += f1_score(train_target, train_pred)
    train_fpr, train_tpr, thresholds = roc_curve(train_target, train_pred)
    train_auc += auc(train_fpr, train_tpr)

    test_pred = []
    dist_1 = abs(cdist(test_data, clf.cluster_centers_)[:, 0])
    dist_2 = abs(cdist(test_data, clf.cluster_centers_)[:, 1])
    for j in range(len(dist_1)):
        if dist_1[j] < dist_2[j]:
            test_pred.append(cluster1_label)
        else:
            test_pred.append(cluster2_label)
    test_accuracy += accuracy_score(test_target, test_pred)
    test_precision += precision_score(
        test_target, test_pred)
    test_recall += recall_score(test_target, test_pred)
    test_f1_score += f1_score(test_target, test_pred)
    test_fpr, test_tpr, thresholds = roc_curve(test_target, test_pred)
    test_auc += auc(test_fpr, test_tpr)

print('avg train accuracy =', train_accuracy/30)
print('avg train precision =', train_precision/30)
print('avg train recall =', train_recall/30)
print('avg train f1_score =', train_f1_score/30)
print('avg train auc =', train_auc/30)
print('Confusion Matrix of training in last run ')
print(confusion_matrix(train_target, train_pred))

print('avg test accuracy =', test_accuracy/30)
print('avg test precision =', test_precision/30)
print('avg test recall =', test_recall/30)
print('avg test f1_score =', test_f1_score/30)
print('avg test auc =', test_auc/30)
print('Confusion Matrix of test in last run ')
print(confusion_matrix(test_target, test_pred))

plt.plot(train_fpr, train_tpr, label='train ROC curve')
plt.plot(test_fpr, test_tpr, label='test ROC curve')
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

## iv. Spectral Clustering: Repeat 1(b)iii using spectral clustering, which is clustering based on kernels. Research what spectral clustering is. Use RBF kernel with gamma=1 or find a gamma for which the two clutsres have the same balance as the one in original data set (if the positive class has p and the negative class has n samples, the two clusters must have p and n members). Do not label data based on their proximity to cluster center, because spectral clustering may give you non-convex clusters . Instead, use fit_predict method.

```python
from sklearn.cluster import SpectralClustering

data = pd.read_csv("./src/all_data.csv", header=None)

train_accuracy = 0
test_accuracy = 0
train_precision = 0
test_precision = 0
train_recall = 0
test_recall = 0
train_f1_score = 0
test_f1_score = 0
train_auc = 0
test_auc = 0
for i in range(30):
    train_data, test_data, train_target, test_target = train_test_split(normalize(data.iloc[:, 1:]), data.iloc[:, 0],
                                                                        test_size=0.2, stratify=data.iloc[:, 0])
    clf = SpectralClustering(n_clusters=2, affinity="rbf", n_init=50, gamma=1)
    train_pred = clf.fit_predict(train_data)

    cluster1_index = np.where(train_pred == 0)
    cluster2_index = np.where(train_pred == 1)

    cluster1_label = np.argmax(np.bincount(
        train_target.iloc[cluster1_index]))
    cluster2_label = np.argmax(np.bincount(
        train_target.iloc[cluster2_index]))
    train_pred[cluster1_index] = cluster1_label
    train_pred[cluster2_index] = cluster2_label

    train_accuracy += accuracy_score(train_target, train_pred)
    train_precision += precision_score(
        train_target, train_pred)
    train_recall += recall_score(train_target, train_pred)
    train_f1_score += f1_score(train_target, train_pred)
    train_fpr, train_tpr, thresholds = roc_curve(train_target, train_pred)
    train_auc += auc(train_fpr, train_tpr)

    test_pred = clf.fit_predict(test_data)

    cluster1_index = np.where(test_pred == 0)
    cluster2_index = np.where(test_pred == 1)

    cluster1_label = np.argmax(np.bincount(
        test_target.iloc[cluster1_index]))
    cluster2_label = np.argmax(np.bincount(
        test_target.iloc[cluster2_index]))
    test_pred[cluster1_index] = cluster1_label
    test_pred[cluster2_index] = cluster2_label

    test_accuracy += accuracy_score(test_target, test_pred)
    test_precision += precision_score(
        test_target, test_pred)
    test_recall += recall_score(test_target, test_pred)
    test_f1_score += f1_score(test_target, test_pred)
    test_fpr, test_tpr, thresholds = roc_curve(test_target, test_pred)
    test_auc += auc(test_fpr, test_tpr)

print('avg train accuracy =', train_accuracy/30)
print('avg train precision =', train_precision/30)
print('avg train recall =', train_recall/30)
print('avg train f1_score =', train_f1_score/30)
print('avg train auc =', train_auc/30)
print('Confusion Matrix of training in last run ')
print(confusion_matrix(train_target, train_pred))

print('avg test accuracy =', test_accuracy/30)
print('avg test precision =', test_precision/30)
print('avg test recall =', test_recall/30)
print('avg test f1_score =', test_f1_score/30)
print('avg test auc =', test_auc/30)
print('Confusion Matrix of test in last run ')
print(confusion_matrix(test_target, test_pred))

plt.plot(train_fpr, train_tpr, label='train ROC curve')
plt.plot(test_fpr, test_tpr, label='test ROC curve')
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

# (b) Repeat each of the following two procedures 50 times. You will have 50 errors for 90 SVMs per each procedure

## i. Train a SVM with a pool of 10 randomly selected data points from the training set using linear kernel and L1 penalty. Select the penalty parameter using 10-fold cross validation.2 Repeat this process by adding 10 other randomly selected data points to the pool, until you use all the 900 points. Do NOT replace the samples back into the training set at each step. Calculate the test error for each SVM. You will have 90 SVMs that were trained using 10, 20, 30, ... , 900 data points and their 90 test errors. You have implemented passive learning.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score, auc
import warnings

warnings.filterwarnings("ignore")

t = pd.read_csv("./src/train2.csv", header=None)
test = pd.read_csv("./src/test2.csv", header=None)
test_data = test.iloc[:, :-1]
test_target = test.iloc[:, -1]
SVCParameters = {"C": np.logspace(-3, 7, 10)}
passive_test_errors = []
for i in range(50):
    test_error_series = []
    new_train_data = pd.DataFrame()
    new_train_target = pd.DataFrame()
    train = t.copy()
    for j in range(90):
        sample10 = train.sample(n=10, axis=0)
        train.drop(index=sample10.index, inplace=True)
        new_train_data = new_train_data.append(sample10.iloc[:, :-1])
        new_train_target = pd.concat(
            [new_train_target, sample10.iloc[:, -1]], axis=0, ignore_index=True)
        clf = LinearSVC(penalty='l1', dual=False)
        CV = GridSearchCV(clf, SVCParameters, cv=5, n_jobs=-1)
        CV.fit(new_train_data, new_train_target.values.ravel())
        best_clf = CV.best_estimator_
        test_error_series.append(1-CV.score(test_data, test_target))
    passive_test_errors.append(test_error_series)
plt.figure(figsize=(16, 9))
for i in passive_test_errors:
    plt.plot(range(0, 900, 10), i, marker='',
             color=np.random.rand(3,), linewidth=1, alpha=0.3)
plt.xlabel('The number of sampled data')
plt.ylabel('Error rate')
plt.show()

```

## ii. Train a SVM with a pool of 10 randomly selected data points from the training set3 using linear kernel and L1 penalty. Select the parameters of the SVM with 10-fold cross validation. Choose the 10 closest data points in the training set to the hyperplane of the SVM4 and add them to the pool. Do not replace the samples back into the training set. Train a new SVM using the pool. Repeat this process until all training data is used. You will have 90 SVMs that were trained using 10, 20, 30,..., 900 data points and their 90 test errors. You have implemented active learning.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score, auc
import warnings

warnings.filterwarnings("ignore")

t = pd.read_csv("./src/train2.csv", header=None)
test = pd.read_csv("./src/test2.csv", header=None)
test_data = test.iloc[:, :-1]
test_target = test.iloc[:, -1]
SVCParameters = {"C": np.logspace(-3, 7, 10)}
active_test_errors = []
for i in range(50):
    train = t.copy()
    test_error_series = []
    new_train_data = pd.DataFrame()
    new_train_target = pd.DataFrame()
    train_data, sample10_data, train_target, sample10_target = train_test_split(train.iloc[:, :-1], train.iloc[:, -1],
                                                                                test_size=10, stratify=train.iloc[:, -1])
    new_train_data = new_train_data.append(sample10_data)
    new_train_target = pd.concat(
        [new_train_target, sample10_target], axis=0, ignore_index=True)
    train.drop(index=sample10_target.index, inplace=True)
    train = train.reset_index(drop=True)
    clf = LinearSVC(penalty='l1', dual=False)
    CV = GridSearchCV(clf, SVCParameters, cv=5)
    CV.fit(new_train_data, new_train_target.values.ravel())
    best_clf = CV.best_estimator_

    for j in range(89):
        sample10_index = abs(best_clf.decision_function(
            train.iloc[:, :-1])).argsort()[:10]
        new_train_data = new_train_data.append(train.iloc[sample10_index, :-1])
        new_train_target = pd.concat(
            [new_train_target, train.iloc[sample10_index, -1]], axis=0, ignore_index=True)
        train.drop(index=sample10_index, inplace=True)
        train = train.reset_index(drop=True)
        clf = LinearSVC(penalty='l1', dual=False)
        CV = GridSearchCV(clf, SVCParameters, cv=5)
        CV.fit(new_train_data, new_train_target.values.ravel())
        best_clf = CV.best_estimator_
        test_error_series.append(1-CV.score(test_data, test_target))
    active_test_errors.append(test_error_series)
plt.figure(figsize=(16, 9))
for i in active_test_errors:
    plt.plot(range(10, 900, 10), i, marker='',
             color=np.random.rand(4,), linewidth=1, alpha=0.3)
plt.xlabel('The number of sampled data')
plt.ylabel('Error rate')
plt.show()
```

Plot average test error versus number of training instances for both active and passive learners on the same figure and report your conclusions. Here, you are actually obtaining a learning curve by Monte-Carlo simulation

```python
plt.figure(figsize=(16, 9))
plt.plot(range(0, 900, 10), np.average(passive_test_errors,axis=0), marker='',
             color=np.random.rand(4,), linewidth=2, alpha=0.3,label="passive learning")
plt.plot(range(10, 900, 10),np.average(active_test_errors,axis=0), marker='',
             color=np.random.rand(4,), linewidth=2, alpha=0.3,label="active learning")
plt.xlabel('The number of sampled data')
plt.ylabel('Average Error rate')
plt.title("Learning Curve")
plt.legend()
plt.show()
```
